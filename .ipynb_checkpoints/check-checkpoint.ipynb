```python
import numpy as np

data = np.load('test_muon_0-1gev_1k.npz')

ranges = data['ranges'] #naive
pls = data['path_lengths'] #detailed
```

```python
for i, pl in enumerate(pls):
    if ranges[i] > pl:
        print('-------------')
        print('naive greater than detailed!')
        print(f'event {i}')
        print(f'naive: {ranges[i]} mm')
        print(f'detailed: {pl} mm')
```
